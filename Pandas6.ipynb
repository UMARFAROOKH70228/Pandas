```python
#CREATING DATAFRAME WITH MISSING VALUES
```

```python
raw_data = {'first_name':['Amar','Akbar','Anthony','Umar',np.nan,np.nan],
         'last_name':['Patil',np.nan,'Paul','Nadaf','More',np.nan],
         'age':[23,np.nan,36,24,73,np.nan],
         'sex':['m',np.nan,'m','f','f',np.nan],
         'PreTestscore':[4,np.nan,np.nan,62,70,np.nan],
         'PostTestscore':[25,np.nan,np.nan,2,3,np.nan]}
df = pd.DataFrame(raw_data,columns=['first_name','last_name','age','sex','PreTestscore','PostTestscore'])
df
        
```

```python
#DROPPING MISSING OBSERVATION or DELETING ROWS WITH NAN
```

```python
df_no_missing = df.dropna()
df_no_missing
```

```python
#IF ROW WITH ALL NAN'S THEN ONLY ITS DELETE 
```

```python
df_cleaned = df.dropna(how='all')
df_cleaned
```

```python
#CREATE NEW COLUMN FULL OF NAN VALUES 
```

```python
df['location'] = np.nan
df
```

```python
#DROP COLOUMN IF THEY CONTAIN MISSING VALUES/NAN VALUES 
```

```python
df.dropna(axis=1,how='all')
df.fillna(0)
df
```

```python
#FILL IN MISSING IN PRETESTSCORE WITH MEAN VALUE OF PRE TEST SCORE
#INPLACE=TRUE: THAT MEANS THE CHANGES ARE SAVED TO THE DF RIGHTAWAT PERMANANTLY
```

```python
df["PreTestscore"].fillna(df["PreTestscore"].mean(),inplace=True)
df
```

```python
#FILL IN MISSSING IN POSTTESTSCORE WITH EACH SEX MEAN VALUE OF POST TEST SCORE
```

```python
df["PostTestscore"].fillna(df.groupby("sex")["PostTestscore"].transform("mean"),inplace=True)
```

```python
df
```

```python
#SELECT SOME ROWS BUT IGNORE THE MISSING DATA POINTS1
```

```python
df[df['age'].notnull()&df['sex'].notnull()]
```

```python
df
```

```python
#BACKFILL & FFILL
```

```python
df.fillna(method='bfill')
```

```python
df.fillna(method='ffill',axis=1)
```

```python
#HIERARCHICAL INDEXING
```

```python
raw_data = {'regiment':['Nightwalks','Nightwalks','Nightwalks','Nightwalks','Dragons','Dragons','Dragons','Dragons','MF','MF','MF','MF'],
            'company' :['1st','2nd','1st','1st','2nd','1st','2nd','1st','1st','2nd','1st','2nd'],
            'name' :['miller','jackson','Ali','miller','cooze','jackon','sone','solom','piger','raint','rag','rag'],
            'pretestscore' :[4,2,3,2,4,2,3,2,5,6,1,5],
            'posttestscore' :[25,42,25,45,12,45,78,65,12,10,14,56]}
df = pd.DataFrame(raw_data,columns = ['regiment','company','name','pretestscore','posttestscore'])
df
```

```python
#SET THE HIERCHICAL INDEX BUT LEAVE THE COLUMN IN PLACE 
```

```python
df.set_index(['regiment','company'],drop=False)  #FROP=FALSE MEANS KEEP THE DATA
df
```

```python
#TO SAVE THE DATA 
```

```python
df.set_index(['regiment','company'],inplace=True)
df
```

```python
#SWAP THE LEVELS IN THE INDEX
```

```python
df.swaplevel('regiment','company')
```

```python
#SUMMERIZE THE RESULTS BY REGIMENT
```

```python
df.sum(level='regiment')
```
